# Task3: Price Range vs. Online Delivery and Table Booking

```python
# Question:- Analyze if there is a relationship between the price range and the availability of online delivery and table booking.
from scipy.stats import chi2_contingency, f_oneway
# If 'Price Range' is categorical, we can use value counts
# Group by 'Online Delivery' and 'Table Booking' for the analysis

# Check the distribution of Price Range with Online Delivery
delivery_group = df.groupby('Has Online delivery')['Price range'].mean()  # i can use 'median' here
print("Average Price Range with Online Delivery:")
print(delivery_group)

# Check the distribution of Price Range with Table Booking
booking_group = df.groupby('Has Table booking')['Price range'].mean()  # i can use 'median' here
print("\nAverage Price Range with Table Booking:")
print(booking_group)

# --- Visualization: Boxplot for Price Range and Delivery ---
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Has Online delivery', y='Price range', hue='Has Online delivery', palette='Set2', legend=False)

plt.title('Price Range vs Online Delivery')
plt.xlabel('Online Delivery Available')
plt.ylabel('Price Range')
plt.show()

# --- Visualization: Boxplot for Price Range and Table Booking ---
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Has Table booking', y='Price range', hue='Has Table booking', palette='Set2', legend=False)
plt.title('Price Range vs Table Booking')
plt.xlabel('Table Booking Available')
plt.ylabel('Price Range')
plt.show()

# --- Statistical Tests ---
# Chi-square test for independence (if 'Price Range', 'Online Delivery', and 'Table Booking' are categorical)
# Create contingency tables (categorical data)
contingency_delivery = pd.crosstab(df['Has Online delivery'], df['Price range'])
contingency_booking = pd.crosstab(df['Has Table booking'], df['Price range'])

# Chi-square test for Online Delivery and Price Range
chi2_delivery, p_delivery, dof, expected_delivery = chi2_contingency(contingency_delivery)
print(f"\nChi-square test for Online Delivery vs Price Range: p-value = {p_delivery:.4f}")

# Chi-square test for Table Booking and Price Range
chi2_booking, p_booking, dof, expected_booking = chi2_contingency(contingency_booking)
print(f"\nChi-square test for Table Booking vs Price Range: p-value = {p_booking:.4f}")

# ANOVA test (if 'Price Range' is numeric) to compare across groups
anova_delivery = f_oneway(
    df[df['Has Online delivery'] == 'Yes']['Price range'],
    df[df['Has Online delivery'] == 'No']['Price range']
)
print(f"\nANOVA test for Online Delivery vs Price Range: p-value = {anova_delivery.pvalue:.4f}")

anova_booking = f_oneway(
    df[df['Has Table booking'] == 'Yes']['Price range'],
    df[df['Has Table booking'] == 'No']['Price range']
)
print(f"ANOVA test for Table Booking vs Price Range: p-value = {anova_booking.pvalue:.4f}")
```

```python
#Question:- Determine if higher-priced restaurants are more likely to offer these services.
from scipy.stats import chi2_contingency, f_oneway
# Example structure: Price range (1-4), Has Online delivery (Yes/No), Has Table booking (Yes/No)
# --- Group Analysis ---
# Mean Price range by Online Delivery
delivery_group = df.groupby('Has Online delivery')['Price range'].mean()
print("Average Price Range with Online Delivery:")
print(delivery_group)

# Mean Price range by Table Booking
booking_group = df.groupby('Has Table booking')['Price range'].mean()
print("\nAverage Price Range with Table Booking:")
print(booking_group)

# --- Visualizations ---
# Boxplot for Price Range vs Online Delivery
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Has Online delivery', y='Price range', palette='Set2')
plt.title('Price Range vs Online Delivery')
plt.xlabel('Online Delivery Available')
plt.ylabel('Price Range')
plt.show()

# Boxplot for Price Range vs Table Booking
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Has Table booking', y='Price range', palette='Set2')
plt.title('Price Range vs Table Booking')
plt.xlabel('Table Booking Available')
plt.ylabel('Price Range')
plt.show()

# --- Statistical Tests ---
# Chi-square test for independence (treating Price range as categorical)
contingency_delivery = pd.crosstab(df['Has Online delivery'], df['Price range'])
chi2_delivery, p_delivery, dof, expected_delivery = chi2_contingency(contingency_delivery)
print(f"\nChi-square test for Online Delivery vs Price Range: p-value = {p_delivery:.4f}")

contingency_booking = pd.crosstab(df['Has Table booking'], df['Price range'])
chi2_booking, p_booking, dof, expected_booking = chi2_contingency(contingency_booking)
print(f"Chi-square test for Table Booking vs Price Range: p-value = {p_booking:.4f}")

# ANOVA test (treating Price range as numeric)
anova_delivery = f_oneway(
    df[df['Has Online delivery'] == 'Yes']['Price range'],
    df[df['Has Online delivery'] == 'No']['Price range']
)
print(f"\nANOVA test for Online Delivery vs Price Range: p-value = {anova_delivery.pvalue:.4f}")

anova_booking = f_oneway(
    df[df['Has Table booking'] == 'Yes']['Price range'],
    df[df['Has Table booking'] == 'No']['Price range']
)
print(f"ANOVA test for Table Booking vs Price Range: p-value = {anova_booking.pvalue:.4f}")

# --- Optional: Proportion of restaurants offering services by Price range ---
delivery_proportion = df.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack()
booking_proportion = df.groupby('Price range')['Has Table booking'].value_counts(normalize=True).unstack()
print("\nProportion of Restaurants Offering Online Delivery by Price Range:")
print(delivery_proportion)
print("\nProportion of Restaurants Offering Table Booking by Price Range:")
print(booking_proportion)
```
